```python
# 합계값을 계산하는 알고리즘

데이터들의 합계를 구하는 가장 단순한 알고리즘은 sum 이지만

arr[0] + arr[1] + ...

로도 표현 가능하다. 하지만 갯수가 늘어나면 사용하기 어렵다.
먼저 반복 구조로 변형하자

arr[0] -> sum
sum + arr[1] -> sum
sum + arr[2] -> sum
...

```

```python
# 단순 알고리즘
arr = [12,13,11,14,10]

sum = arr[0]
sum = sum + arr[1]
sum = sum + arr[2]
sum = sum + arr[3]

print(sum)

# 반복 알고리즘
sum = 0
i = 0
while i < len(arr):
    sum += arr[i]
    i += 1
print(sum)
```

```python
# 최대값을 찾는 알고리즘
# 첨자 index     0   1  2   3  4
# 데이터         12 13 11 14 10

arr = [12, 13, 11, 14, 10]

max = arr[0]

for i in arr:
  if i > max:
    max = i

print(max)
```

```python
# 유명한 알고리즘
# 기본적인 처리 절차를 가지고 있는 좋은 알고리즘이다
# 좋은 프로그램을 만들기 위한 사고방식이나 힌트가 많이 포함되어 있다
# 유명한 알고리즘을 학습하는 것은 프로그래밍 기술 향상에 커다란 도움이 된다

# 탐색
# Linear Search 선형 탐색법 - 앞에서부터 순서대로 찾는다
# Binary Search 이진 탐색법 - 범위를 절반식 줄여서 찾는다
# Hash Search   해시 탐색법 - 검색 즉시 찾는다. 단 저장할때 가공해서 저장한다

# 정렬
# Selection Sort 단순 선택법 - 최소/최대값을 선택하여 앞/뒤부터 순서대로 나열한다
# Bubble Sort    단순 교환법 - 옆에있는 데이터를 교환하면서 자리를 바꾼다
# Insertion Sort 단순 삽입법 - 데이터를 올바른 위치에 삽입하면서 자리를 바꾼다
# Quict Sort     퀵정렬      - 기준 데이터를 기반으로 대소 분할을 반복하여 자리를 바꾼다

# 에라토스테네스의 체        - 소수 구하는 알고리즘
# 유클리드 알고리즘          - 최대공약수 구하는 알고리즘
```

```python
# Linear Search 선형 탐색법

# 보관된 데이터를 맨 앞부터 순서대로 원하는 값을 찾을 때까지 탐색한다. 
# 탐색 처리는 반복 구조를 사용하여 기술한다. 
# 반복 구조에서는 종료 조건을 잊지 말자. 

첨자 (index)   0  1  2  3  4
데이터          4  2  3  5  1    찾을 데이터  :  5
    
arr = [12, 13, 11, 14, 10]
i=0

while (i<len(arr)):
    if arr[0] = 0:
        print("{}번째 요소에 있습니다".format(i))
        break
```

```python
# try except 구문

try:
  num_input = input('정수 입력 >')

  num_a = int(num_input)

  print('원의 반지름 : ', num_a)
  print('원의 둘레', 2 * 3.14 * num_a)
  print('원의 넓이', 3.14 * num_a * num_a)
  
except:
  print('정수를 입력하세요')
```

```python
try:
# 예외가 발생할 가능성이 있는 코드

except:
# 예외가 발생했을 때 실행할 코드

else:
# 예외가 발생하지 않았을 때 실행할 코드 

finally:
# 항상 실행 시킬 코드

# try + except
# try + except + else
# try + except + finally
# try + except + else + finally
# try + finally
```

```python
def test():
  print('test()함수의 첫 줄입니다.')
  try:
    print('try  구문이 실행되었습니다.')
    return
    print('try 구문의 return 아래입니다.')
  except:
    print('except구문이 실행되었습니다.')
  else:
    print('else구문이 실행되었습니다.')
  finally:
    print('finally 구문이 실행되었습니다.')
  print('test()함수의 마지막 줄입니다.')

test()
```
