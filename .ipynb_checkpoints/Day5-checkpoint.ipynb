```python
1. 알고리즘
- 문제나 과제를 해겨하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어 또는 생각
- 알고리즘 자체는 일상생활에서도 이미 많이 사용되고 있다
- 요리의 레시피, 음악의 악보, 가전제품 사용설명서 등
- 문자, 기호, 일러스트 등으로 나타낸다.

2. 알고리즘 프로그램
- 알고리즘을 프로그래밍 언어로 기술하면 프로그램이 된다.
- 프로그래밍 언어는 컴퓨터에 지시하기 위한 인공 언어이다.

3. 알고리즘 작성
- 프로그램 작성은 기획, 설계(*), 프로그래밍, 디버그로 진행한다
- 알고리즘은 설계 단계에서 필요하다

4. 좋은 알고리즘
- 알기쉽다.(이해하기 좋다)
- 속도가 빠르다
- 효율적이다.(메모리를 적게 사용한다)
- 재이용이 쉽다(객체지향 프로그래밍)

5. 알고리즘 필요성
- 좋은 프로그램을 만들기 위해 (빠르고, 효율적, 범용성)
- 프로그램의 좋고 나쁨을 판단하기 위해
- 프로그램 작성 과정 전체를 효율화 하기 위해
- 프로그램 기술을 향상시키기 위해

6. 알고리즘 조건
- 정확한 결과를 얻을 수 있어야 한다 (알고리즘은 문제나 과제를 해결하기 위한 절차 )
- 반드시 종료되어야 한다 (중간 처리나 절차가 잘못 만들어지면 영원히 끝나지 않는다 )

7. 알고리즘의 기본형 3가지
- 순차 구조 : 처음부터 순서대로 처리하는 절차
  실행을 원하는 처리를 위로부터 선서대로 작성한다
```

```python
# 가변 매개 변수
# 함수를 선언할 때 매개변수와 함수를 호출할때 매개변수가 같아야만 했다.

def print_n_times(n,*val):
    for i in range(n):
        for v in val:
            print(v)
        print()
print_n_times(3,"hi","hello","world")
```

```python
def sum_all(*val):
    val = list(val)
    sum = 0
    for i in val:
        sum += i
    print(sum)
sum_all(1,2,3,5)
```

```python
def test(a,b=10,c=100):
    print(a+b+c)

# 1. 기본형태
test(10,20,30)

# 1. 생략형태
test(10)

# 1. 순서 섞어서 만든 형태
test(c = 10, a = 30, b = 40)

# 1. 일부 지정 형태
test(10,c=30)
```

```python
def sum_all(start,end):
    sum = 0
    for n in range(start,end+1):
        sum += n
    return sum
print("0에서 100까지의 합 :",sum_all(0,100))
print("0에서 1000까지의 합 :",sum_all(0,1000))
print("50에서 100까지의 합 :",sum_all(50,100))
print("500에서 1000까지의 합 :",sum_all(500,1000))
```

```python
# 반복문으로 팩토리얼 구하기

def facto(n):
    if n == 1:     
        return 1    
    return n * facto(n - 1)    
 
print(facto(5))
```

```python
# input 과 output 있는 함수
def add(a,b):
    return a+b

# input 없고 output 있는 함수
def say():
    return "hi"

# input 있고 output 없는 함수
def add(a,b):
    print(a+b)
    # return 없으면 output 없는 것이다
    
# input 과 output 없는 함수
def say():
    print("hi")
    
```

```python
def add(a,b):
    return a+b

print(add(a=3,b=7))
print(add(b=7,a=3))
```

```python
def add_many(*args):
    result = 0
    for arg in args:
        result += arg
    return result

add_many(3)
add_many(55,321,2)
add_many(2,2,2,3,3,3,4,54,5,6)
```
