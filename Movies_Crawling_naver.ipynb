```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
html = requests.get("https://movie.naver.com/movie/sdb/rank/rmovie.naver")
```

```python
soup = bs(html.text)
```

```python
len(soup.select('td.title'))
```

```python
movies = soup.select('td.title')
```

```python
movies[0]
```

```python
movies[0].select('a')[0].text
```

```python
data = []
rank = 1

for movie in movies:
    title = movie.select('a')[0].text
    data.append([rank,title])
    rank += 1
    
data
```

```python
df = pd.DataFrame(data, columns = ['순위','제목'])
df
```

```python
df.to_excel('movies_naver.xlsx', index=False)
```
